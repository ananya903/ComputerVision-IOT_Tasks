# Object Detection / Optical Character Recognition (ORC) 

● Implement an object detector which identifies the classes of the objects in 
an image or video. OR

● Character detector which extracts printed or handwritten text from an 
image or video.



```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
i = 'hello.jpg'
```

```python
reader = easyocr.Reader(['en'])
result = reader.readtext(i)
result
```

```python
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img = cv2.imread(i)
spacer = 100
for detection in result: 
    top_left = tuple(detection[0][0])
    bottom_right = tuple(detection[0][2])
    text = detection[1]
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
    img = cv2.putText(img,text,(20,spacer), font, 0.5,(0,255,0),2,cv2.LINE_AA)
    spacer+=15
    
plt.imshow(img)
plt.show()
```
